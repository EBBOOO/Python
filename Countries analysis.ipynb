```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
```

```python
us = data[data.country == 'United States']
china = data[data.country == 'China']
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
plt.title('GPD per Capita')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.legend(['U.S.', 'China'])
plt.show()
```

```python
us_growth = us.gdpPerCapita/us.gdpPerCapita.iloc[0] *100
china_growth = china.gdpPerCapita/china.gdpPerCapita.iloc[0] *100
```

```python
plt.plot(us.year, us_growth)
plt.plot(china.year, china_growth)
plt.title('GPD per Capita Growth (first year = 100)')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita Growth')
plt.legend(['U.S.', 'China'])
plt.show()
```

```python
china_population_growth = china.population/china.population.iloc[0] * 100
us_population_growth = us.population/us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.title('population')
plt.xlabel('Year')
plt.ylabel('population')
plt.legend(['U.S.', 'China'])
plt.show()
```

```python
plt.plot(us.year, us_population_growth)
plt.plot(china.year, china_population_growth)
plt.legend(['U.S.', 'China'])
plt.xlabel('Year')
plt.ylabel('Population Growth')
plt.show()
```

```python
obama = pd.read_csv('obama.csv', parse_dates=['year_month'])
```

```python
plt.plot(obama.year_month, obama.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

```python
obama_mean = obama.groupby('year_month').mean()
obama_median = obama.groupby('year_month').median()
obama_25 = obama.groupby('year_month').quantile(0.25)
obama_75 = obama.groupby('year_month').quantile(0.75)

plt.plot(obama_mean.index, obama_mean.approve_percent, 'red')
plt.plot(obama_median.index, obama_median.approve_percent, 'green')
plt.plot(obama_25.index, obama_25.approve_percent, 'blue')
plt.plot(obama_75.index, obama_75.approve_percent, 'black')

plt.plot(obama.year_month, obama.approve_percent, 'o', markersize=2, alpha=0.3)
plt.legend(['mean','median', '25th percentile', '75th percentile'])
plt.show()
```
