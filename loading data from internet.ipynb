```python
#!pip install yfinance
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import urllib
import matplotlib.dates as mdates
```

```python
#def graph_data(stock):
#    stock_prices_url = 'http://chartapi.finance'

#data = yf.download('TSLA', '2010-01-01','2019-01-01')

apple = yf.download('AAPL', '2010-01-01','2019-01-01')
google = yf.download('GOOG', '2010-01-01','2019-01-01')
microsoft = yf.download('MSFT', '2010-01-01','2019-01-01')

google['Date'] = google.index
apple['Date'] = apple.index
microsoft['Date'] = microsoft.index

fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

ax1.fill_between(google.Date, google.Close,0, color='g',label='Google')
ax1.fill_between(apple.Date, apple.Close,0, color='r',label='Apple')
ax1.fill_between(microsoft.Date, microsoft.Close,0,color='b',label='Microsoft')

#ax1.plot(google.Close, label='Google', color='g')
#ax1.plot(apple.Close, label='Apple', color='r')
#ax1.plot(microsoft.Close, label='Microsoft', color='b')
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
ax1.grid()
#data.Close.plot(label='Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stocks')
plt.legend(loc='upper left')
plt.show()
```

```python
twitter = yf.download('TWTR', '2010-01-01','2019-01-01')

twitter['Date'] = twitter.index

#fig = plt.figure()
ax2 = plt.subplot2grid((1,1),(0,0))

ax2.fill_between(twitter.Date, twitter.Close, twitter.Close[0],where=(twitter.Close > twitter.Close[0]), color='g', alpha=0.5, label='Gain')
ax2.fill_between(twitter.Date, twitter.Close, twitter.Close[0],where=(twitter.Close < twitter.Close[0]), color='r', alpha=0.5, label='Loss')

ax2.plot(twitter.Close, label='Twitter', color='b')
for label in ax2.xaxis.get_ticklabels():
    label.set_rotation(45)
ax2.grid()

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stocks')
plt.legend(loc='upper left')
plt.show()
```
